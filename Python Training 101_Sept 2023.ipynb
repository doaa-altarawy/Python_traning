# Part 3: plotting

```python
df['Sales'].plot()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date')['Sales'].plot()
```

```python
df.set_index('Order Date').resample('m')['Sales'].sum().plot(title='Sales per year');
```

```python
df.set_index('Order Date').resample('y')['Sales'].sum().plot.bar(title='Sales per year', rot=45);
```

```python
df.groupby('Sub-Category')['Sales'].sum().plot.barh();
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values().plot.barh(title='Total sales per sub-category');
plt.savefig('total_sales_per_car.png')
```

```python
df.groupby('Category')['Quantity'].sum().sort_values().plot.barh(title='Total Quantity sold  per category');
```
